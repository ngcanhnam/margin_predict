```python





# --- FLOW RATIOS (month windows) ---
# create safe denominators using fillna(0)+eps

df_base = df_base.fillna(0)

df_base['flow_ratio_3m']  = (df_base['total_buy_3m']) / (df_base['total_sell_3m'])
df_base['flow_ratio_12m'] = (df_base['total_buy_12m']) / (df_base['total_sell_12m'])
df_base['flow_ratio_2w']  = (df_base['total_buy_2w']) / (df_base['total_sell_2w'])

# -- Buy/Sell Presure --- 
df_base['buy_sell_pressure_3m']  = (df_base['total_buy_3m'] - df_base['total_sell_3m']) / df_base['total_trading_amount_3m']
df_base['buy_sell_pressure_12m'] = (df_base['total_buy_12m'] - df_base['total_sell_12m']) / df_base['total_trading_amount_12m']

# -- order ratio ---
df_base['order_buy_ratio_3m']  = df_base['order_buy_3m'] / (df_base['order_buy_3m'] + df_base['order_sell_3m'])
df_base['order_buy_ratio_12m'] = df_base['order_buy_12m'] / (df_base['order_buy_12m'] + df_base['order_sell_12m'])
df_base['order_sell_ratio_3m']  = df_base['order_sell_3m'] / (df_base['order_buy_3m'] + df_base['order_sell_3m'])
df_base['order_sell_ratio_12m'] = df_base['order_sell_12m'] / (df_base['order_buy_12m'] + df_base['order_sell_12m'])

## -- total_trading_ratio --- 

df_base['total_buy_ratio_3m']  = df_base['total_buy_3m'] / df_base['total_trading_amount_3m']
df_base['total_buy_ratio_12m'] = df_base['total_buy_12m'] / df_base['total_trading_amount_12m']
df_base['total_sell_ratio_3m']  = df_base['total_sell_3m'] / df_base['total_trading_amount_3m']
df_base['total_sell_ratio_12m'] = df_base['total_sell_12m'] / df_base['total_trading_amount_12m']






# --- TURNOVER RATIOS (theo tháng) ---

df_base['turnover_3m']  = df_base['total_trading_amount_3m']  / (df_base['avg_nav_ck_3m']  + df_base['avg_nav_cash_3m'])
df_base['turnover_12m'] = df_base['total_trading_amount_12m'] / (df_base['avg_nav_ck_12m'] + df_base['avg_nav_cash_12m'])


#(Tỷ lệ tiền mặt/Tổng tài sản):

df['Cash_Drag_Ratio_3m'] = df['avg_nav_cash_3m'] / (df['avg_nav_ck_3m'] + df['avg_nav_cash_3m'])
df['Cash_Drag_Ratio_12m'] = df['avg_nav_cash_12m'] / (df['avg_nav_ck_12m'] + df['avg_nav_cash_12m'])



# (Bùng nổ giao dịch)
#df['Trading_Intensity_Surge'] = df['trading_count_3m']/(df['trading_count_12m'] / 4)
df_base['Trading_Intensity_Surge'] = df_base['trading_count_3m'] / df_base['trading_count_12m']
df_base['Speculative_Index'] = df_base['turnover_3m'] / df_base['turnover_12m']

df_base['asset_allocation_stock_ratio_3m'] = df_base['avg_nav_ck_3m'] / (df_base['avg_nav_ck_3m'] + df_base['avg_nav_cash_3m'])
df_base['asset_allocation_stock_ratio_12m'] = df_base['avg_nav_ck_12m'] / (df_base['avg_nav_ck_12m'] + df_base['avg_nav_cash_12m'])
#df['net_flow_trend'] = df['net_cash_flow_3m'] - (df['net_cash_flow_12m'] / 4)

#df['order_buy_sell_pressure_3m'] = df['order_buy_3m'] / (df['order_sell_3m'] + 1e-6)
#df['is_loss_chasing'] = ((df['profit_ratio_3m'] < 0) & (df['trading_count_3m'] > df['trading_count_12m']/4)).astype(int)


df_base['net_buy_pressure_3m'] = (df_base['order_buy_3m'] - df_base['order_sell_3m']) / (df_base['order_buy_3m'] + df_base['order_sell_3m'])


```

```python
import pandas as pd
import numpy as np

# 1. Nhóm các biến chỉ có giá trị dương (Cần chặn trên 99%)
pos_cols = [
    'turnover_12m', 'flow_ratio_12m',
    'flow_ratio_3m','order_buy_ratio_3m', 'order_buy_ratio_12m', 'order_sell_ratio_3m', 'order_sell_ratio_12m',
    'Trading_Intensity_Surge', 'login_trend', 'asset_allocation_stock_ratio_3m', 'asset_allocation_stock_ratio_12m', 'trading_count_12m',
    'trading_count_3m', "buy_intensity_3m"
]

# 2. Nhóm các biến có giá trị âm (Cần chặn cả 2 đầu: 1% và 99%)
# Đặc biệt quan trọng cho Beta, Alpha, Profit, Net Flow
two_way_cols = [
    'profit_ratio_3m',
    'cash_flow_3m', 'cash_flow_12m', 
    'buy_sell_pressure_3m', 'buy_sell_pressure_12m',
    'order_buy_sell_pressure_3m', 'beta_2w', 'alpha_2w', 'drawdown_2w', 'Cash_Drag_Ratio_3m', 'Cash_Drag_Ratio_12m', #'buy_intensity_negative_3m'
]

def handle_outliers(df):
    df_capped = df.copy()
    
    # Xử lý nhóm chặn trên 0.99
    for col in pos_cols:
        if col in df_capped.columns:
            upper_limit = df_capped[col].quantile(0.99)
            df_capped[col] = df_capped[col].clip(upper=upper_limit)
            
    # Xử lý nhóm chặn 2 đầu (0.01 và 0.99)
    for col in two_way_cols:
        if col in df_capped.columns:
            lower_limit = df_capped[col].quantile(0.01)
            upper_limit = df_capped[col].quantile(0.99)
            df_capped[col] = df_capped[col].clip(lower=lower_limit, upper=upper_limit)
            
    return df_capped

# Áp dụng
df_base = handle_outliers(df_base)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ==============================================================================
# 1. DATA PREPARATION (ĐÃ SỬA LỖI THIẾU CỘT)
# ==============================================================================

# --- Bước A: Lọc danh sách ID của 2 nhóm ---
# Nhóm FP: Model báo 1 - Thực tế 0
ids_fp = df_final[(df_final['pred'] == 1) & (df_final['loan_label_next'] == 0)]['master_account']

# Nhóm TP: Model báo 1 - Thực tế 1 (Active Margin)
ids_tp = df_final[(df_final['pred'] == 1) & (df_final['loan_label_next'] == 1)]['master_account']

# --- Bước B: Lấy dữ liệu Buy Intensity từ df_base ---
# Giả sử cột 'buy_intensity_3m' nằm ở df_base (hoặc bạn đổi thành df_final nếu cột đó nằm ở đó)
# Chỉ lấy những cột cần thiết để vẽ
cols_needed = ['master_account', 'buy_intensity_3m']
source_df = df_base[cols_needed].copy() 

# --- Bước C: Tạo Dataframe cho nhóm FP ---
df_fp = source_df[source_df['master_account'].isin(ids_fp)].copy()
df_fp['Group_Label'] = 'Potential Leads (FP)'

# --- Bước D: Tạo Dataframe cho nhóm TP ---
df_tp = source_df[source_df['master_account'].isin(ids_tp)].copy()
# Sample nhóm TP cho bằng số lượng FP (để biểu đồ cân đối)
if len(df_tp) > len(df_fp):
    df_tp = df_tp.sample(n=len(df_fp), random_state=42)
df_tp['Group_Label'] = 'Active Margin (TP)'

# --- Bước E: Ghép và Làm sạch ---
df_viz_fp = pd.concat([df_fp, df_tp])

# Lọc nhiễu Intensity (giữ < 6)
# Lúc này df_viz_fp CHẮC CHẮN đã có cột buy_intensity_3m
df_viz_fp = df_viz_fp[df_viz_fp['buy_intensity_3m'] < 6]

# ==============================================================================
# 2. SETUP STYLE & PALETTE (Giữ nguyên)
# ==============================================================================
custom_palette_fp = {
    'Potential Leads (FP)': '#F39C12', 
    'Active Margin (TP)': '#C0392B'
}
sns.set_theme(style="whitegrid", rc={"grid.color": ".92", "axes.edgecolor": ".8"})

# ==============================================================================
# 3. PLOTTING FUNCTION (Giữ nguyên)
# ==============================================================================
def plot_fp_validation(data, variable='buy_intensity_3m'):
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # Plot 1: Box Plot
    sns.boxplot(ax=axes[0], data=data, x='Group_Label', y=variable, 
                palette=custom_palette_fp, linewidth=1.5, width=0.5, showfliers=False)
    axes[0].set_title(f"Box Plot: Behavior Similarity\n(Leads vs Active Users)", fontsize=14, fontweight='bold', color='#2C3E50')
    axes[0].set_ylabel("Buy Intensity (3M)")

    # Plot 2: Histogram
    sns.histplot(ax=axes[1], data=data, x=variable, hue='Group_Label', 
                 palette=custom_palette_fp, kde=False, element="step", stat="density", common_norm=False, alpha=0.25)
    axes[1].set_title(f"Histogram: Volume Pattern Overlay", fontsize=14, fontweight='bold', color='#2C3E50')
    
    # Plot 3: KDE Plot
    sns.kdeplot(ax=axes[2], data=data, x=variable, hue='Group_Label', 
                fill=True, common_norm=False, palette=custom_palette_fp, alpha=0.3, linewidth=2.5, clip=(0, 6))
    axes[2].set_title(f"KDE: Are Leads behaving like Users?", fontsize=14, fontweight='bold', color='#2C3E50')
    
    plt.suptitle(f" VALIDATING FALSE POSITIVES (N={len(data):,})", 
                 fontsize=18, fontweight='bold', y=1.05, color='#E67E22')
    sns.despine(left=True)
    plt.tight_layout()
    plt.show()

# Chạy hàm vẽ
plot_fp_validation(df_viz_fp, variable='buy_intensity_3m')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# ==============================================================================
# 1. DATA PREPARATION (FULL POPULATION - NO SAMPLING)
# ==============================================================================

# --- Bước A: Lọc ID ---
ids_fp = df_final[(df_final['pred'] == 1) & (df_final['loan_label_next'] == 0)]['master_account']
ids_tp = df_final[(df_final['pred'] == 1) & (df_final['loan_label_next'] == 1)]['master_account']

# --- Bước B: Lấy dữ liệu Drawdown từ df_base ---
# Đảm bảo lấy đúng cột drawdown_2w
cols_needed = ['master_account', 'drawdown_2w']
source_df = df_base[cols_needed].copy() 

# --- Bước C: Tạo Dataframe cho nhóm FP (Potential Leads) ---
df_fp = source_df[source_df['master_account'].isin(ids_fp)].copy()
df_fp['Group_Label'] = 'Potential Leads (FP)'

# --- Bước D: Tạo Dataframe cho nhóm TP (Active Margin) ---
df_tp = source_df[source_df['master_account'].isin(ids_tp)].copy()
df_tp['Group_Label'] = 'Active Margin (TP)'

# --- Bước E: Ghép và Làm sạch ---
df_viz_dd = pd.concat([df_fp, df_tp])

# Lọc nhiễu: Drawdown thường là số âm (từ -1 đến 0) hoặc số dương thể hiện độ sụt giảm.
# Giả sử trong data của bạn: 0 là không giảm, -0.3 là giảm 30%.
# Ta lọc vùng từ -0.5 (giảm 50%) đến 0.05 (số dư nhiễu dương nhẹ)
df_viz_dd = df_viz_dd[pd.to_numeric(df_viz_dd['drawdown_2w'], errors='coerce').between(-0.5, 0.05)]

print(f"Tổng số lượng Potential Leads (FP): {len(df_fp):,}")
print(f"Tổng số lượng Active Margin (TP): {len(df_tp):,}")

# ==============================================================================
# 2. SETUP STYLE
# ==============================================================================
custom_palette_fp = {
    'Potential Leads (FP)': '#F39C12', # Vàng Cam
    'Active Margin (TP)': '#C0392B'    # Đỏ Đậm
}
sns.set_theme(style="whitegrid", rc={"grid.color": ".92", "axes.edgecolor": ".8"})

# ==============================================================================
# 3. PLOTTING FUNCTION
# ==============================================================================
def plot_drawdown_validation(data, variable='drawdown_2w'):
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # --- Plot 1: Box Plot (Mức độ sụt giảm trung bình) ---
    sns.boxplot(
        ax=axes[0], data=data, x='Group_Label', y=variable, 
        palette=custom_palette_fp, linewidth=1.5, width=0.5, showfliers=False
    )
    # Kẻ đường 0 (Bảo toàn vốn)
    axes[0].axhline(0, color='#27AE60', linestyle='--', linewidth=1.5)
    axes[0].text(0.5, 0.01, 'Preserved Capital (0%)', ha='center', fontsize=9, color='#27AE60', fontweight='bold')
    
    axes[0].set_title(f"Box Plot: Loss Magnitude\n(How deep is the pain?)", fontsize=14, fontweight='bold', color='#2C3E50')
    axes[0].set_ylabel("Drawdown 2W (%)")
    # Đảo trục Y nếu muốn nhìn kiểu "Thác đổ" (tùy chọn), ở đây giữ nguyên âm dưới dương trên
    
    # --- Plot 2: Histogram (Phân phối tần suất) ---
    sns.histplot(
        ax=axes[1], data=data, x=variable, hue='Group_Label', 
        palette=custom_palette_fp, kde=False, element="step", 
        stat="density", common_norm=False, alpha=0.25
    )
    axes[1].axvline(0, color='#27AE60', linestyle='--', linewidth=1.5)
    axes[1].set_title(f"Histogram: Portfolio Resilience", fontsize=14, fontweight='bold', color='#2C3E50')
    axes[1].set_xlabel("Drawdown 2W")
    
    # --- Plot 3: KDE Plot (Vùng Nguy Hiểm) ---
    sns.kdeplot(
        ax=axes[2], data=data, x=variable, hue='Group_Label', 
        fill=True, common_norm=False, palette=custom_palette_fp, 
        alpha=0.3, linewidth=2.5, clip=(-0.5, 0.05)
    )
    
    # Annotation: Vùng "Gồng Lỗ" (Drawdown < -15%)
    danger_threshold = -0.15
    axes[2].axvline(danger_threshold, color='#C0392B', linestyle=':', linewidth=2)
    
    # Tô màu vùng nguy hiểm cho nhóm Leads
    if len(axes[2].get_lines()) > 0:
        kde_line = axes[2].get_lines()[0]
        x_data = np.array(kde_line.get_data()[0])
        y_data = np.array(kde_line.get_data()[1])
        # Tô màu vùng bên trái ngưỡng nguy hiểm
        axes[2].fill_between(x_data, y_data, where=(x_data <= danger_threshold), color='#7F8C8D', alpha=0.2)

    axes[2].text(-0.16, axes[2].get_ylim()[1]*0.5, "STRESS ZONE\n(Drop > 15%)\n=> Need Average Down", 
                 color='#C0392B', fontweight='bold', ha='right', fontsize=10)
    
    axes[2].set_title(f"KDE: Who holds 'Bags' better?", fontsize=14, fontweight='bold', color='#2C3E50')
    axes[2].set_xlabel("Drawdown 2W")
    
    # --- Final Touches ---
    plt.suptitle(f" VALIDATING PORTFOLIO HEALTH (Drawdown 2W - N={len(data):,})", 
                 fontsize=18, fontweight='bold', y=1.05, color='#E67E22')
    sns.despine(left=True)
    plt.tight_layout()
    plt.show()

# Chạy hàm vẽ
plot_drawdown_validation(df_viz_dd, variable='drawdown_2w')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# ==============================================================================
# 1. CHUẨN BỊ DỮ LIỆU CHUNG
# ==============================================================================
# Lấy đủ các trường cần thiết
cols_needed = ['master_account', 'beta_2w', 'buy_intensity_3m', 'drawdown_2w', 'alpha_2w']
source_df = df_base[cols_needed].copy()

ids_fp = df_final[(df_final['pred'] == 1) & (df_final['loan_label_next'] == 0)]['master_account']
ids_tp = df_final[(df_final['pred'] == 1) & (df_final['loan_label_next'] == 1)]['master_account']

df_fp = source_df[source_df['master_account'].isin(ids_fp)].copy()
df_fp['Group_Label'] = 'Potential Leads (FP)'

df_tp = source_df[source_df['master_account'].isin(ids_tp)].copy()
df_tp['Group_Label'] = 'Active Margin (TP)'

df_plot_all = pd.concat([df_fp, df_tp])

# Lọc nhiễu chung
df_plot_all = df_plot_all[
    pd.to_numeric(df_plot_all['beta_2w'], errors='coerce').between(-0.5, 3.5) &
    pd.to_numeric(df_plot_all['buy_intensity_3m'], errors='coerce').between(0, 6) &
    pd.to_numeric(df_plot_all['drawdown_2w'], errors='coerce').between(-0.5, 0.05) &
    pd.to_numeric(df_plot_all['alpha_2w'], errors='coerce').between(-0.25, 0.25)
]

# Random sample để vẽ
df_viz = df_plot_all.sample(frac=1, random_state=42).reset_index(drop=True)
palette_map = {'Potential Leads (FP)': '#F39C12', 'Active Margin (TP)': '#C0392B'}

# ==============================================================================
# 2. HÀM VẼ MA TRẬN CHIẾN LƯỢC
# ==============================================================================
def plot_strategic_matrix(data, x_col, y_col, x_limit, y_limit, x_thresh, y_thresh, 
                          title, xlabel, ylabel, quadrant_texts):
    
    plt.figure(figsize=(14, 9))
    sns.set_theme(style="whitegrid", rc={"grid.color": ".95"})

    # Scatter Plot
    sns.scatterplot(
        data=data, x=x_col, y=y_col, hue='Group_Label', palette=palette_map,
        alpha=0.4, s=20, edgecolor=None, linewidth=0
    )

    # Kẻ trục phân định
    plt.axvline(x_thresh, color='#34495E', linestyle='--', linewidth=1.5)
    plt.axhline(y_thresh, color='#34495E', linestyle='--', linewidth=1.5)

    # Gắn nhãn 4 vùng (Quadrant Text)
    # Cấu trúc quadrant_texts: [(x, y, text, color), ...]
    for (qx, qy, qtext, qcolor) in quadrant_texts:
        plt.text(qx, qy, qtext, color=qcolor, fontsize=11, fontweight='bold', ha='center',
                 bbox=dict(facecolor='white', alpha=0.9, edgecolor=qcolor, boxstyle='round,pad=0.5'))

    # Trang trí
    plt.title(title, fontsize=16, fontweight='bold', color='#2C3E50', y=1.02)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xlim(x_limit)
    plt.ylim(y_limit)

    # Legend Fix
    legend = plt.legend(loc='upper right', frameon=True, title='Phân nhóm')
    for lh in legend.legend_handles: 
        lh.set_alpha(1)
        lh.set_markersize(12)

    sns.despine()
    plt.tight_layout()
    plt.show()

# ==============================================================================
# 3. THỰC THI VẼ 2 KỊCH BẢN
# ==============================================================================

# --- KỊCH BẢN A: TÂM LÝ CHIẾN ĐẤU (Drawdown vs Intensity) ---
# X: Drawdown (Lỗ nhiều <--- 0 ---> Lỗ ít)
# Y: Intensity (Mua mạnh vs Mua yếu)

# Cấu hình Text cho 4 vùng
quadrants_A = [
    # Góc Trái Trên (Lỗ Sâu + Mua Mạnh) -> Đang gồng lỗ/bắt đáy
    (-0.25, 5.0, "FIGHTERS / AVG DOWN\n(Gồng lỗ quyết liệt)\n=> Cần Margin để về bờ", '#C0392B'),
    
    # Góc Phải Trên (Lỗ Ít + Mua Mạnh) -> Đang hưng phấn
    (-0.02, 5.0, "CONFIDENT TRADERS\n(Tự tin/Có lãi)\n=> Upsell Margin gia tăng", '#27AE60'),
    
    # Góc Trái Dưới (Lỗ Sâu + Mua Yếu) -> Sợ hãi
    (-0.25, 0.5, "FEAR / CAPITULATION\n(Sợ hãi/Đông cứng)\n=> Cần tư vấn/CSKH", '#7F8C8D'),
    
    # Góc Phải Dưới (Lỗ Ít + Mua Yếu) -> Thận trọng
    (-0.02, 0.5, "CAUTIOUS\n(Thận trọng/Quan sát)\n=> Chào hàng lãi thấp", '#F39C12')
]

plot_strategic_matrix(
    data=df_viz,
    x_col='drawdown_2w', y_col='buy_intensity_3m',
    x_limit=(-0.4, 0.02), y_limit=(0, 6.5),
    x_thresh=-0.10, y_thresh=1.5, # Ngưỡng: Lỗ -10% và Mua > 1.5
    title='MA TRẬN TÂM LÝ: AI ĐANG "BẮT DAO RƠI"? (Drawdown vs Intensity)',
    xlabel='Drawdown 2W (Mức sụt giảm - Càng trái càng lỗ)',
    ylabel='Buying Intensity (Cường độ mua)',
    quadrant_texts=quadrants_A
)

# --- KỊCH BẢN B: NĂNG LỰC & ĐẲNG CẤP (Beta vs Alpha) ---
# X: Beta (Rủi ro)
# Y: Alpha (Hiệu quả)

quadrants_B = [
    # Góc Trái Trên (Beta thấp + Alpha cao) -> Smart Money
    (0.2, 0.15, "SMART MONEY / VIP\n(Rủi ro thấp - Lãi cao)\n=> KH Mơ ước (Priority 1)", '#27AE60'),
    
    # Góc Phải Trên (Beta cao + Alpha cao) -> Skilled Speculator
    (2.5, 0.15, "SKILLED SPECULATORS\n(Đánh phiêu - Có lãi)\n=> Cấp High Limit", '#F39C12'),
    
    # Góc Phải Dưới (Beta cao + Alpha âm) -> Gamblers
    (2.5, -0.15, "GAMBLERS / RISK\n(Rủi ro cao - Lỗ nặng)\n=> Cảnh báo rủi ro", '#C0392B'),
    
    # Góc Trái Dưới (Beta thấp + Alpha âm) -> Ineffective
    (0.2, -0.15, "INEFFECTIVE\n(An toàn - Nhưng kém)\n=> Cần tư vấn danh mục", '#7F8C8D')
]

plot_strategic_matrix(
    data=df_viz,
    x_col='beta_2w', y_col='alpha_2w',
    x_limit=(-0.5, 3.5), y_limit=(-0.25, 0.25),
    x_thresh=1.0, y_thresh=0.0, # Ngưỡng: Beta=1, Alpha=0
    title='MA TRẬN NĂNG LỰC: SMART MONEY vs DUMB MONEY (Beta vs Alpha)',
    xlabel='Beta (Rủi ro thị trường)',
    ylabel='Alpha (Lợi nhuận vượt trội)',
    quadrant_texts=quadrants_B
)
```
